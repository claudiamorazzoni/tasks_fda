##### Lists and dictionaries

Lists are similar to strings, which are a sequence of characters, however, unlike strings, items in a list can be of different types.

An object is a value to which a variable can refer or point and can be identified using its identifiers, which are unique. We can test whether two names refer to the same object using the is operator. 

```python
# The variable a and b receive the same value
a = "banana"
b = "banana"

print(a is b) # The is operator returns True if the two references are to the same object.
```

This is not the case with lists. Consider the example below. Here "a" and "b" refer to two different lists, each of which happens to have the same elements as values.

```python
a = [81, 82, 83]
b = [81, 82, 83]

print(a is b)
print(a == b)
```

"a" and "b" have the same value but do not apply to the same object. In this case, they take a place (index) in the computer's memory.

A dictionary is a type of data structure of the collection type, that is, an object that contains more than one value. Unlike lists in which elements are accessed through a position or index, in dictionaries, information is accessed using keys. The elements of a dictionary are stored in an unordered manner. They use a pair structure containing key and value "{key:value}", so that their elements are unique, although the elements of a value can be duplicated since the key is different.


```python
a = {"d1":81, "d2":82, "d3":83}
b = {"d1":81, "d2":82, "d3":83}
c = {"d4":81, "d5":82, "d6":83}
d = {"d1":86, "d2":85, "d3":86}

print(a is b)
print(a == b)
print(b == c)
print(d == a)
```

"a" and "b" have the same key but do not apply to the same object. "c" does not have the same key so do not apply to the same object and "d" have the same key but not the same value so do not apply to the same object.

```python
def counts (l):
    """ 
    A function named counts that takes a list as input and returns a dictionary
    """
    # list comprehension using for loop and count function to get the number of occurrences os each item, thus forming the pair that a dictionary requires
    return (dict ((x,l.count(x)) for x in l))
```

##### Finally the dicerolls function 

```python
import random as rd

def dicerolls (k=1, n=10):
    """ 
    A function named dicerolls that simulate randomly rolling dice n times and returns a dictionary with the number of times each possible total face value occurred
    """
    cont = 0  # We need an interactor for the loop count
    d0 = [] # We need an empty list to store the drawn data
    if k==1:
        while True: # The loop just finish when cont >= n
            dn = [d0.append(rd.randint(1,6))] # We need an empty local list to store the data
            if cont < n:
                cont += 1
                if cont >= n:
                    return (dict((x,d0.count(x)) for x in d0)) # # list comprehension to get the number of occurrences os each item, thus forming the pair that a dictionary requires
    elif k>=2:
        t=k*1 # when we are playing with 2 or more dices
        s=k*6 # when we are playing with 2 or more dices
        while True:
            d0.append(rd.randint(t,s)) # t and s to delimit the number of numbers to be drawn
            if cont < n:
                cont += 1
                if cont >= n:
                    return (dict((x,d0.count(x)) for x in d0)) 
    else:
        return print("You can only play with integers numbers") # a exception when the player When the player types 0 or a float number
```

## Task 3 -  Flipping a coin

The idea is to simulate the game of a coin with 50/50 heads and tails and we'll find out how it works if we play 1000 times.

Here we will understand that each of the functions that we write can be used and called by the other functions that we write. This is one of the most important ways that allows computer scientists to take a big problem and break it down into a group of smaller problems. This process of breaking down a problem into smaller subproblems is called functional decomposition.[1]

Here is a simple example of functional decomposition using two functions. The first, called square, simply computes the square of a given number. A second, called square_sum, uses the square function to determine the sum of the square of three numbers.

```python
# The square function returns the result of multiplying the values of x and y
def square(x): 
    y = x * x
    return y

# The squere_sum function returns the result of the addition after the square function is called
def square_sum(x,y,z): # the values of y, x and z at first are -5, 2 and 10
    a = square(x) # calling the square function three time
    b = square(y)
    c = square(z)
    return a+b+c

a = -5
b = 2
c = 10

result = square_sum(a,b,c) # calling the square_sum function
print(result)
```

Although this is a very simple idea, in practice this example illustrates many important concepts in Python, including local and global variables, as explained in the previous task, along with passing parameters.

In this way we will continue with the flipping_a_coin function itself for this task.

```python
# First it will be necessary to import the numpy and random libraries
import numpy as np
import random as rd

# here we declare two variables in the form of an opposite dictionary so that we can capture the number of heads and tails drawn
id_coin = {0: 'tail', 1: 'head'}
coin_id = {'tail': 0, 'head': 1}

# the flippinig_a_coin function returns a number (or zero or one) drawn that will determine whether heads or tails
def  flipping_a_coin():
    return  id_coin[rd.randint(0,1)]

# this function returns the number of heads and tails per round
def playing_n_times(n=10, printer=True):
    global n_heads
    global n_tails
    global results
    
    if printer:
        results=[]
        for i in range(n):
            results.append(coin_id[flipping_a_coin()])
            n_heads = sum(results)
            n_tails = n-n_heads
            
    if printer:
        print(f'Round')
        print(f'heads - {n_heads} times')
        print(f'tails - {n_tails} times')
    return results, n_heads, n_tails



n_heads_round = []
n_tails_round = []

# this function calls the playing_n_times function and has a loop for the number of rounds
def number_round(n_round = 10, play_per_round = 10):
    print(f'Flipping a coin {play_per_round} times in {n_round} rounds\n')
    for rounds in range(n_round):
        _, n_heads, n_tails = playing_n_times(play_per_round, True) # calling the Playing_n_times function
        n_heads_round.append(n_heads) # it acumulates the results from n_heads in playing_n_times function
        n_tails_round.append(n_tails) # it acumulates the results from n_heads in playing_n_times function
       
        
        
    
number_round(1000,15) # calling the number_round function
```

```python
# ploting the results of the heads to represent the resulting list of 1,000 numbers
import matplotlib.pyplot as plt
_=plt.hist(n_heads_round, 100)
plt.title('Frequency of the number of heads')
```

```python
# ploting the results of the tails to represent the resulting list of 1,000 numbers
n, bins, patches = plt.hist(n_tails_round, 100, density=True, facecolor='g', alpha=0.75)
plt.title('Frequency of the number of tails')
```

With this histogram, we arrive here at what we call binomial distribution, very common in the observation of natural elements.

The binomial distribution is the distribution of probability and discrete statistics of the number of successes resulting from a given sequence of attempts, which follow the following characteristics:[2]

- Finite sample space;

- Only two possible outcomes (success or failure) for each attempt;

- All elements must have equal chances of occurrence;

- Events must be independent of each other.
