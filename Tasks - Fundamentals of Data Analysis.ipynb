##### Lists and dictionaries

Lists are similar to strings, which are a sequence of characters, however, unlike strings, items in a list can be of different types.

An object is a value to which a variable can refer or point and can be identified using its identifiers, which are unique. We can test whether two names refer to the same object using the is operator. 

```python
# The variable a and b receive the same value
a = "banana"
b = "banana"

print(a is b) # The is operator returns True if the two references are to the same object.
```

This is not the case with lists. Consider the example below. Here "a" and "b" refer to two different lists, each of which happens to have the same elements as values.

```python
a = [81, 82, 83]
b = [81, 82, 83]

print(a is b)
print(a == b)
```

"a" and "b" have the same value but do not apply to the same object. In this case, they take a place (index) in the computer's memory.

A dictionary is a type of data structure of the collection type, that is, an object that contains more than one value. Unlike lists in which elements are accessed through a position or index, in dictionaries, information is accessed using keys. The elements of a dictionary are stored in an unordered manner. They use a pair structure containing key and value "{key:value}", so that their elements are unique, although the elements of a value can be duplicated since the key is different.


```python
a = {"d1":81, "d2":82, "d3":83}
b = {"d1":81, "d2":82, "d3":83}
c = {"d4":81, "d5":82, "d6":83}
d = {"d1":86, "d2":85, "d3":86}

print(a is b)
print(a == b)
print(b == c)
print(d == a)
```

"a" and "b" have the same key but do not apply to the same object. "c" does not have the same key so do not apply to the same object and "d" have the same key but not the same value so do not apply to the same object.

##### List comprehension

List comprehension is generally used for code optimization and performance. [5]
Suppose we want to add items from one list to another list.

```python
# Creating any list and an empty list
old_list = [1,2,3,4,5,6]
new_list = []

# At first we could use this code o include numbers in the new list (for loop)
for i in old_list:
    new_list.append(i)

print(new_list)
```

##### Def Counts

```python
def counts (l):
    """ 
    A function named counts that takes a list as input and returns a dictionary
    """
    # list comprehension using for loop and count function to get the number of occurrences os each item, thus forming the pair that a dictionary requires
    return (dict ((x,l.count(x)) for x in l))
```

##### Finally the dicerolls function 

```python
import random as rd

def dicerolls (k=1, n=10):
    """ 
    A function named dicerolls that simulate randomly rolling dice n times and returns a dictionary with the number of times each possible total face value occurred
    """
    cont = 0  # We need an interactor for the loop count
    d0 = [] # We need an empty list to store the drawn data
    if k==1:
        while True: # The loop just finish when cont >= n
            dn = [d0.append(rd.randint(1,6))] # We need an empty local list to store the data
            if cont < n:
                cont += 1
                if cont >= n:
                    return (dict((x,d0.count(x)) for x in d0)) # # list comprehension to get the number of occurrences os each item, thus forming the pair that a dictionary requires
    elif k>=2:
        t=k*1 # when we are playing with 2 or more dices
        s=k*6 # when we are playing with 2 or more dices
        while True:
            d0.append(rd.randint(t,s)) # t and s to delimit the number of numbers to be drawn
            if cont < n:
                cont += 1
                if cont >= n:
                    return (dict((x,d0.count(x)) for x in d0)) 
    else:
        return print("You can only play with integers numbers") # a exception when the player When the player types 0 or a float number
```

## Task 3 - Numpy Random Binomial 

```python
import numpy as np
import random as rd


id_coin = {0: 'tail', 1: 'head'}
coin_id = {'tail': 0, 'head': 1}

def  flipping_a_coin():
    return  id_coin[rd.randint(0,1)]
    
def playing_n_times(n=10, printer=True):
    global n_heads
    global n_tails
    global results
    
    if printer:
        print(f'Flipping a coin {n} times\n')
        results=[]
        for i in range(n):
            results.append(coin_id[flipping_a_coin()])
            n_heads = sum(results)
            n_tails = n-n_heads
    if printer:
        print(f'Result')
        print(f'heads - {n_heads} times')
        print(f'tails - {n_tails} times')
    return results, n_heads, n_tails



n_heads_round = []
n_tails_round = []

def number_round(n_round = 10, play_per_round = 10):
    for rounds in range(n_round):
        _, n_heads, n_tails = playing_n_times(play_per_round, True)
        n_heads_round.append(n_heads)
        n_tails_round.append(n_tails)
        
        
    
number_round(1000, 15)
```
